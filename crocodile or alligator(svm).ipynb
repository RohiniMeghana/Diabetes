```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
#we can create 40 separable points
X,y=make_blobs(n_samples=40,centers=2,random_state=2)
```

```python
#fit the model
clf=svm.SVC(kernel='linear',C=1)
clf.fit(X,y)
```

```python
#display the graph
plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=plt.cm.Paired)
plt.show()
```

```python
#using to predict unknown data
newData=[[3,4],[5,6]]
print(clf.predict(newData))
```

```python
#fit the model
clf=svm.SVC(kernel='linear',C=1000)
clf.fit(X,y)
plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=plt.cm.Paired)
#plt.show()
#plot the decision function
ax=plt.gca()
xlim=ax.get_xlim()
ylim=ax.get_ylim()
#create a grid to evaluate the model
xx=np.linspace(xlim[0],xlim[1],30)
yy=np.linspace(ylim[0],ylim[1],30)
YY,XX=np.meshgrid(yy,xx)
xy=np.vstack([XX.ravel(),YY.ravel()]).T
z=clf.decision_function(xy).reshape(XX.shape)
#plot decision boundaries and the margins
ax.contour(XX,YY,z,colors='k',levels=[-1,0,1],alpha=0.5,linestyles=['--','-','--'])
#plot the support vectors
ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100,linewidth=1,facecolors='none')
plt.xlabel("width of the snout")
plt.ylabel("size of the body")
plt.title("Crocodile or Alligator")
plt.show()
```
